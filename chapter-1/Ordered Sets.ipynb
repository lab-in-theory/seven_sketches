## Ordered Sets 

A preorder a $\le$ b, c $\le$  d, e

```julia
using LightGraphs
using TikzGraphs

adjacency_matrix = [ # a b c d e
                       0 1 1 0 0; # a
                       0 0 0 1 0; # b
                       0 0 0 0 1; # c
                       0 0 0 0 0; # d
                       0 0 0 0 0; # e
                   ]

ordered_set = DiGraph(adjacency_matrix)
TikzGraphs.plot(ordered_set,["a","b","c","d","e"])
```
